```python
# Import the necessary libraries
import requests
from bs4 import BeautifulSoup
import sys


# Make a request to the website you want to scrape
# Handle errors related to connection at the same time
page = "https://en.wikipedia.org/wiki/Cyberpunk_2077"

def get_page(page):
    try:
        r = requests.get(page)
    except requests.exceptions.ConnectionError as err:
        print("Connection Error : \n", err)
        sys.exit()
    except requests.exceptions.HTTPError as err:
        print("Http error : \n", err)
        sys.exit()
    return r

r = get_page(page)

# Parse the HTML of the website
# Handle parsing errors
def get_soup(r):
    try:
        soup = BeautifulSoup(r.text, "html.parser")
    except Exception as err:
        print("Error while parsing :\n", err)
        sys.exit()
    return soup

# To handle eg. Rate Limiting, we can use the time library to retry requests
# Captcha solving services such as Buster are also available 

soup = get_soup(r)

# Find the element(s) you want to extract
elements = soup.find_all("p")

# Extract the text from the element
element_text = elements[2].text

print(element_text)
```

The below code downloads the images from the page. It is specifically made to fit wikipedia image storage


```python
import os
import urllib.request

# Create the "images" folder if it doesn't exist
if not os.path.exists("images"):
    os.makedirs("images")

# Find all the <img> elements on the page
pg_images = soup.find_all("img")

for one_image in pg_images:
    image_url = one_image.get('src')

    if ".svg" in image_url:
        continue
    if "thumb" not in image_url:
        continue
    image_name = image_url.split("px-")[-1]
    image_url = image_url.split("//")[1]
    #print(image_url)
    urllib.request.urlretrieve("http://"+image_url, "images/" + image_name)
```
